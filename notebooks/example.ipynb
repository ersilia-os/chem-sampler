## FasmiFra Sampler
FasmiFra is a fragment-based generative model.

_Before running it, make sure to install fasmifra in the conda environment using opam_

```python
# use a single sampler 
from chemsampler.samplers.fasmifra.sampler import FasmifraSampler
onesampler = FasmifraSampler()
onesampler_results = onesampler.sample(smiles, n=100)
onesampler_results
```

## Bimodal Sampler

If search_pre_calculated=True, it queries a database of 10k generated molecules, whereas if search_pre_calculated=False, it is used in generative mode using a pre-trained model. Bimodal environment needs to be set up to run the generative steps.

```python
# use a single sampler 
from chemsampler.samplers.bimodal.sampler import BimodalSampler
onesampler = BimodalSampler()
onesampler_results = onesampler.sample(smiles, n=100, search_pre_calculated=False,)
onesampler_results
```

## Moler Sampler
If search_pre_calculated=True, it queries a database of 100k generated molecules, whereas if search_pre_calculated=False, it is used in generative mode using a pre-trained model. Moler environment needs to be set up to run the generative steps.

```python
# use a single sampler 
from chemsampler.samplers.moler.sampler import MolerSampler
onesampler = MolerSampler()
onesampler_results = onesampler.sample(smiles, n=100, search_pre_calculated=False)
onesampler_results
```
