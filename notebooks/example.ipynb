```python
from chemsampler import ChemSampler, example
smiles = example() #provides 10 approved drugs as example smiles
smiles
```

### ChEMBL Sampler

Similarity search on the ChEMBL database

```python
cs = ChemSampler(samplers_list= ["ChemblSampler"])
results = cs.sample(smiles, num_samples=100)
results
```

### PubChem Sampler
Similarity search on the PubChem database

```python
cs = ChemSampler(samplers_list= ["PubChemSampler"])
results = cs.sample(smiles, num_samples=100)
results
```

### SmallWorld Sampler
Similarity search on the SmallWorld database

```python
cs = ChemSampler(samplers_list= ["SmallWorldSampler"])
results = cs.sample(smiles, num_samples=100)
results
```

### Stoned Sampler
Generative model that takes the input molecules in batch and introduces small mutations to generate similar molecules

```python
cs = ChemSampler(samplers_list= ["StonedSampler"])
results = cs.sample(smiles, num_samples=100)
results
```

### Moler Sampler
Moler generates new candidates using a pretrained model. Fine-tuning to your own dataset can be also performed (see development section of the README file)

```python
cs = ChemSampler(samplers_list= ["MolerSampler"])
results = cs.sample(smiles, num_samples=100)
results
```

### Bimodal Sampler
Bimodal generates new candidates using a pretrained model. Fine-tuning to your own dataset can be also performed (see development section of the README file)

```python
cs = ChemSampler(samplers_list= ["BimodalSampler"])
results = cs.sample(smiles, num_samples=100)
results
```

### Fasmifra Sampler
Generates new molecules based on fragment assembly. _Before running it, make sure to install fasmifra in the conda environment using opam_

```python
cs = ChemSampler(samplers_list= ["FasmifraSampler"])
results = cs.sample(smiles, num_samples=100)
results
```

### Mollib Sampler
Generates new molecules based on the Mollib package. 

_Note: Mollib will require >30 min to generate molecules_

```python
cs = ChemSampler(samplers_list= ["MollibSampler"])
results = cs.sample(smiles, num_samples=100)
results
```
