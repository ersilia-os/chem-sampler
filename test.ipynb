```python
from chemsampler.descriptors.descriptor import DescriptorCalculator

def _calculate_input_descriptors(input_smiles, descriptor_id):
    dc = DescriptorCalculator(model_id=descriptor_id)
    descs = dc.calculate([input_smiles])[0]
    return list(descs)

def _calculate_sampled_descriptors(sampled_smiles, descriptor_id):
    dc = DescriptorCalculator(model_id=descriptor_id)
    print(sampled_smiles)
    descs = dc.calculate(sampled_smiles)
    return descs
```

```python
descriptor_id = "eos5axz"
origin_descs = _calculate_input_descriptors(input_smiles, descriptor_id)
sampled_descs = _calculate_sampled_descriptors(sampled_smiles, descriptor_id)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import DataStructs

def calculate_tanimoto_similarity(fp1, fp2):
    return DataStructs.TanimotoSimilarity(fp1, fp2)

def np_to_bv(fv):
    bv = DataStructs.ExplicitBitVect(len(fv))
    for i,v in enumerate(fv):
        if v:
            bv.SetBit(i)
    return bv

similarities = [calculate_tanimoto_similarity(np_to_bv(origin_descs), np_to_bv(sampled_desc)) for sampled_desc in sampled_descs]
```
